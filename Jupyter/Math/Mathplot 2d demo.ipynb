# 2dPlot i Python

```python
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
x=np.arange(0,math.pi*3, 0.05)
y=np.sin(x)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title("Sine wave")
ax.set_xlabel("Angle")
ax.set_ylabel('Sine')
plt.show()
```

```python
x = np.array([0,13])
y = np.array([0, 250])
plt.plot(x,y)
plt.show()
```
