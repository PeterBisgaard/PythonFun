# Parametric equations
Demo from https://www.youtube.com/watch?v=OoVLt_2wJlk

## Import numpy library

```python
import numpy as np
```

## Parameters
pointsRangeInterval from 0 to 10*pi with 1000 data points -> https://numpy.org/doc/stable/reference/generated/numpy.linspace.html

```python
radius = 1
pointsRangeInterval = np.linspace(0,10*np.pi, 1000)
```

## Parametric equations in cartesian co-ordinates
cartesian co-ordinates = normal x,y,z co-ordinates

```python
x = radius*np.cos(pointsRangeInterval)
y = radius*np.sin(pointsRangeInterval)
z = pointsRangeInterval
```

## Import plotting libraries
mpl_toolkits is the tool for 3D plots -> https://matplotlib.org/2.1.2/mpl_toolkits/index.html

```python
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as Axes3D
```

## Adjustment for figure
111 = the position of the plot on canvas, See https://stackoverflow.com/questions/3584805/in-matplotlib-what-does-the-argument-mean-in-fig-add-subplot111

```python
fig = plt.figure('Parametic curves')
ax = fig.add_subplot(111, projection='3d')
```

## Plot the data
'-r' = line type and color = solid line red. See https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.plot.html

```python
ax.plot(x,y,z,'-r', linewidth=3)
```

## Add lables

```python
ax.set_xlabel('X', fontweight = 'bold', fontsize = 14)
ax.set_ylabel('Y', fontweight = 'bold', fontsize = 14)
ax.set_zlabel('Z', fontweight = 'bold', fontsize = 14)
```

## Add title

```python
plt.title('Parametric Curve', fontweight = 'bold', fontsize=16)
```

## Show the figure

```python
plt.show()
```

```python
import numpy as np
radius = 1
pointsRangeInterval = np.linspace(0,10*np.pi, 1000)
# Parametric equations in cartesian co-ordinates
x = radius*np.cos(pointsRangeInterval)
y = radius*np.sin(pointsRangeInterval)
z = pointsRangeInterval
# Plot setup
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as Axes3D
fig = plt.figure('Parametic curves')
ax = fig.add_subplot(111, projection='3d')
ax.plot(x,y,z,'-r', linewidth=3)
# Add lables
ax.set_xlabel('X', fontweight = 'bold', fontsize = 14)
ax.set_ylabel('Y', fontweight = 'bold', fontsize = 14)
ax.set_zlabel('Z', fontweight = 'bold', fontsize = 14)
plt.title('Parametric Curve', fontweight = 'bold', fontsize=16)
plt.show()
```

### Torus  (plural tori, colloquially donut or doughnut)
Another example, formula from https://en.wikipedia.org/wiki/Parametric_equation#Parametric_surfaces

combined with this example https://www.youtube.com/watch?v=O8NkzKSv7Jw and https://en.wikipedia.org/wiki/Torus

np.meshgrid(angleA,angleB) = https://www.udacity.com/blog/2021/10/numpy-np-meshgrid-tutorial-for-beginners.html

ax.plot_surface(x,y,z,rstride=1, cstride=1, color='Purple') = https://matplotlib.org/stable/gallery/mplot3d/subplot3d.html and https://matplotlib.org/3.5.0/tutorials/toolkits/mplot3d.html for rstride, cstride

```python
import numpy as np
r = 4
R = 10
# Parametric equations in cartesian co-ordinates
angleA = np.linspace(0,2*np.pi, 30) # Top-bund
angleB = np.linspace(0,2*np.pi, 30) # højere-venstre

u,t = np.meshgrid(angleA,angleB)
x = np.cos(t)*(R + r*np.cos(u))
y = np.sin(t)*(R + r*np.cos(u))
z = r*np.sin(u)

# Plot setup
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as Axes3D
fig = plt.figure('Parametic curves')
#ax = fig.add_subplot(111, projection='3d')
ax = plt.axes(projection='3d')
ax.set_xlim3d(-10,10)
ax.set_ylim3d(-10,10)
ax.set_zlim3d(-10,10)
ax.plot_surface(x,y,z, rstride = 1, cstride = 1, color='Purple')


plt.show()
```
